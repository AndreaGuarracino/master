```python
import pandas as pd
import epitopepredict as ep
from epitopepredict import base, sequtils, plotting
```

```python
path = "data/Common_COV_SARS2_N_comparison.fasta"
df = sequtils.fasta_to_dataframe(path)

display(df)
```

```python
print(base.predictors)
```

```python
# Make a choice about which alleles to include
#alleles = base.get_preset_alleles('broad_coverage_mhc1')
alleles = ['HLA-A01:01']

# Make a choice of which predictors to use
predictors = ['mhcnuggets', 'mhcflurry', 'netmhcpan']

outputs = {}

for predictor in predictors:
    predictor = base.get_predictor(predictor)
    outputs[predictor] = predictor.predict_sequences(df, alleles=alleles)
```
