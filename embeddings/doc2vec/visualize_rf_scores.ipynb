```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(context='paper', style='whitegrid', palette='deep', font='serif', font_scale=2, color_codes=True, rc=None)
```

```python
results_csv = "data/results_k4.csv"
```

```python
df = pd.read_table(results_csv, sep=",", index_col = 0)
df
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))
ax = sns.heatmap(df, vmin = 0, vmax = 1, annot=True, fmt=".3", cmap='mako')
ax.set_ylabel("embedding size")
ax.set_xlabel("epochs")
ax.set_title("Robinson-Foulds ratio\nw2v k 4 tree vs. clustalOmega tree")

plt.savefig("/Users/francescoballesio/Desktop/corona/final_plots/best_embedding_w2v_k4.pdf")
plt.savefig("/Users/francescoballesio/Desktop/corona/final_plots/best_embedding_w2v_k4.png", dpi=600)
```
